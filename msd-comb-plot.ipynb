```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import *
from matplotlib.backends.backend_pdf import PdfPages
import glob

filenames = sorted(glob.glob('./with_Fld/*.dat'))
#leg = f[15:20]  
for f in filenames:
    data = np.loadtxt(fname = f)
    tau = range(1,20)
    x = data[:,1]
    plt.plot(tau, x,label=f)
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
#plt.savefig('msd_comb.pdf')
plt.show()

for f in filenames:
    data = np.loadtxt(fname = f)
    tau = range(1,20)
    y = data[:,2]
    plt.plot(tau, y,label=f)
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
#plt.savefig('msd_comb.pdf')
plt.show()

for f in filenames:
    data = np.loadtxt(fname = f)
    tau = range(1,20)
    z = data[:,3]
    plt.plot(tau, z,label=f)
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
#plt.savefig('msd_comb.pdf')
plt.show()
```

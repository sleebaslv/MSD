```python
import numpy as np
import matplotlib.pyplot as plt
data = np.loadtxt('3892.xyz')
time = data[:,0]
x = data[:,1] #x cordinate
y = data[:,2] #y cordinate
z = data[:,3] #z cordinate
fig, axs = plt.subplots(1,3, figsize=(12,6))
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.5, hspace=None)
axs[0].plot(time, x, 'k')
axs[0].set_xlabel('Time(ps)')
axs[0].set_ylabel('X')
axs[0].set_title('X-cordinate')

axs[1].plot(time, y, 'k')
axs[1].set_xlabel('Time(ps)')
axs[1].set_ylabel('Y')
axs[1].set_title('Y-cordinate')

axs[2].plot(time, z, 'k')
axs[2].set_xlabel('Time(ps)')
axs[2].set_ylabel('Z')
axs[2].set_title('Z-cordinate')

plt.show()
```

```python
import sys
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt
import glob

def func(x, p1, p2):
    return ((p1**2)*(x**2))+2*p2*x

data = np.loadtxt('../output/msd20.dat')
#data_f = np.loadtxt('../output/msd_time_avg/msd_field_20.dat')
tau = data[:,0]
msd_x = data[:,1]
msd_y = data[:,2]
msd_z = data[:,3]

vel = []
dc = []

for i in range(1,3):
        popt, pcov = opt.curve_fit(func, tau, data[:,i])
        vel.append(popt[0])
        dc.append(popt[1])
        #print popt[0]
        #print popt[1]
        #slope.append(popt[0])
        #constant.append(popt[1])
        plt.plot(tau, data[:,i])
        plt.plot(tau, func(tau, popt[0], popt[1]))
        plt.show()

vel = np.array(vel)
dc = np.array(dc)

vel = vel.flatten()
dc = dc.flatten()
print vel
print dc

#np.savetxt('../output/diff_coeff-100.dat', [slope], fmt='%.4f', delimiter=' ')

                                                                               
```
